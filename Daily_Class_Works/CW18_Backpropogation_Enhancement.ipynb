### Question 1:
According to the derivation of the Optimal Learning Rate $\eta$ using the Taylor Series expansion (up to the second order), why does $\eta$ need to be recalculated at every iteration $k$?

A) Because the gradient $\nabla J$ is constant, but the Hessian $H$ changes.  
B) Because both the gradient $\nabla J$ and the Hessian $H$ are functions of the current weights $w^k$.  
C) Because the Taylor series only works if the learning rate decreases over time.  
D) Because $\eta$ must always be equal to the inverse of the loss function.  

Answer: (B): The optimal learning rate is derived by minimizing the second-order Taylor approximation: $J(w^{k+1}) \approx J(w^k) - \eta(\nabla J)^T \nabla J + \frac{\eta^2}{2}(\nabla J)^T H (\nabla J)$. Taking the derivative with respect to $\eta$ and setting it to zero gives $\eta^* = \frac{(\nabla J)^T \nabla J}{(\nabla J)^T H (\nabla J)}$. Since both $\nabla J$ and $H$ depend on the current weights $w^k$, the optimal learning rate must be recalculated at each iteration as the weights change.

### Question 2:
In Newton's Method, if the Hessian matrix $H$ has negative eigenvalues, what is a common strategy to ensure the update remains a descent step?

A) Increase the learning rate $\eta$ to skip over the negative curvature.  
B) Use only the first-order gradient and ignore the Hessian entirely.  
C) Regularize the Hessian by adding a term $\alpha I$, effectively blending Newton's method with Gradient Descent.  
D) Flip the sign of the gradient $\nabla J$.  

Answer: (C): When the Hessian has negative eigenvalues, the Newton step $H^{-1}\nabla J$ may point in a direction of ascent rather than descent (the Hessian is not positive semi-definite). Damped Newton's method or Levenberg-Marquardt regularization adds a regularization term: $H' = H + \alpha I$. This ensures $H'$ becomes positive definite, guaranteeing a descent step. It provides a smooth transition between Newton's method (when $\alpha$ is small) and Gradient Descent (when $\alpha$ is large).

### Question 3:
Given a simple 1D objective function $J(w) = w^4$. Suppose at iteration $k$, the current weight is $w^k = 1$.

a) Calculate the Gradient $\nabla J$ at $w=1$.  
b) Calculate the Hessian $H$ at $w=1$.  
c) Using the Newton's Update formula $w^{k+1} = w^k - H^{-1} \nabla J$, calculate the new weight $w^{k+1}$. 

### Answer: 

a) Gradient at $w = 1$:

$$\nabla J = \frac{dJ}{dw} = \frac{d(w^4)}{dw} = 4w^3$$

At $w = 1$: $\nabla J = 4(1)^3 = 4$

gradient is the first derivative of the loss function, representing the direction and magnitude of steepest ascent.

---

b) Hessian at $w = 1$:

$$H = \frac{d^2J}{dw^2} = \frac{d(4w^3)}{dw} = 12w^2$$

At $w = 1$: $H = 12(1)^2 = 12$

The Hessian is the second derivative, representing the curvature of the loss landscape. In 1D, it's a scalar.

---

c) Newton's Update:

$$w^{k+1} = w^k - H^{-1} \nabla J = 1 - \frac{1}{12} \cdot 4 = 1 - \frac{4}{12} = 1 - \frac{1}{3} = \frac{2}{3}$$

Newton's method uses second-order information to make a more informed step than gradient descent. The update $H^{-1}\nabla J$ essentially scales the gradient by the inverse curvature, accounting for the local geometry. Here, the large positive Hessian (H=12) indicates strong curvature, so Newton's method takes a smaller, well-calibrated step (1/3) compared to what a gradient descent would take with a fixed learning rate.
